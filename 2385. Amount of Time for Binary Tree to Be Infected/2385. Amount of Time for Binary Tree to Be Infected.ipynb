1.1 Creating Vertex Class

```python
class Vertex:
        def __init__(self, val):
            self.val = val
            self.next = None
            self.color = None
            self.distace = None
            self.parent = None

        def setNext(self, next):
            self.next = next
        
        
```

1.2 Creating add Edge fuction.

```python
def addEdge(adj, index, val): #taking index and val 
    
    if (index == None or val == None): # <-- There can't be any edge to and from a Null Node
        return
    else:
        if adj.get(index) is None: # <-- We may have visited this index in our adjacency list first time
            adj[index] = Vertex(val)
            return
        else:
            currNode = adj[index] # <-- There may already exsist a LinkedList Node in this index of the adjacency list
            while (currNode.next != None): # <-- Traverse the linked list to the very last node
                currNode = currNode.next
            currNode.setNext(Vertex(val)) # <-- Add the new node as the next of the last node
```

```python
class Queue:
    def __init__(self):
        self.items = []
    
    def is_empty(self):
        return len(self.items) == 0
    
    def enqueue(self, item):
        self.items.append(item) # adding items at the verry end
    
    def dequeue(self):
        if self.is_empty():
            raise IndexError("Dequeue from empty queue")
        return self.items.pop(0)
    
    def peek(self):
        if self.is_empty():
            raise IndexError("Peek form empty queue")
        return self.items[0]
    
    def size(self):
        return len(self.items)
    
    
```

```python
from collections import deque

def bfs(graph, start):
    WHITE, GRAY, BLACK = 'WHITE', 'GRAY', 'BLACK'

    # Initialize vertex properties
    color = {u: WHITE for u in graph}
    distance = {u: float('inf') for u in graph}
    predecessor = {u: None for u in graph}

    # Initialize source vertex properties
    color[start] = GRAY
    distance[start] = 0
    predecessor[start] = None

    # Initialize queue
    queue = deque([start])

    # BFS algorithm adapted for linked list in adjacency list
    while queue:
        u = queue.popleft()
        current_vertex = graph[u]
        while current_vertex:
            v = current_vertex.val
            if color[v] == WHITE:
                color[v] = GRAY
                distance[v] = distance[u] + 1
                predecessor[v] = u
                queue.append(v)
            current_vertex = current_vertex.next
        color[u] = BLACK

    return color, distance, predecessor

```
