2225. Find Players With Zero or One Losses

You are given an integer array matches where matches[i] = [winneri, loseri] indicates that the player winneri defeated player loseri in a match.

Return a list answer of size 2 where:

answer[0] is a list of all players that have not lost any matches.
answer[1] is a list of all players that have lost exactly one match.
The values in the two lists should be returned in increasing order.

Note:

You should only consider the players that have played at least one match.
The testcases will be generated such that no two matches will have the same outcome.

start -> make two list winners and losers -> winner = match[i][0] and loser = match[i][1] -> compare the two
-> see how many time winners are there in the losers ->
answer[0] is a list of players with zero losses -> answer[1] is a list with 1 game loss.
-> result is in ascending order 

More detailed 


```python
class Solution(object):
    def findWinners(self, matches):
        """
        :type matches: List[List[int]]
        :rtype: List[List[int]]
        """
        player_losses = {}

        for winner, loser in matches:
            
            player_losses[loser] = player_losses.get(loser, 0) + 1
            if winner not in player_losses:
                player_losses[winner] = 0

        players_with_zero_losses = [player for player, losses in player_losses.items() if losses == 0]
        players_with_one_loss = [player for player, losses in player_losses.items() if losses == 1]

                
        players_with_zero_losses.sort()
        players_with_one_loss.sort()

        
        answer = [players_with_zero_losses, players_with_one_loss]
        return answer


```
